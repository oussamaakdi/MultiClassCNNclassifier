# Data Cleaning

```python
data_directory = 'data'
image_exts = ['jpeg','jpg','bmp','png']
#tf.keras.utils.image_dataset_from_directory('data',batch_size=32,image_size=(256,256))
```

```python
for image_class in os.listdir(data_directory):
    for image in os.listdir(os.path.join(data_directory,image_class)):
        image_path = os.path.join(data_directory,image_class,image)
        try:
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts:
                print('Image not in extentions list {}'.format(image_path))
                os.remove(image_path)      
        except Exception as e :
                print("Issue with image{}".format(image_path))
                os.remove(image_path)
```

# Creating the data set

```python
dataset =  tf.keras.utils.image_dataset_from_directory('data')
num_classes = len(dataset.class_names)
dataset = dataset.map(lambda x, y: (x, tf.one_hot(y, depth=num_classes)))
```

```python
data_iterator = dataset.as_numpy_iterator()
```

```python
batch =  data_iterator.next()
```

```python
fig , ax =  plt.subplots(ncols=4 , figsize=(20,20))
for idx ,img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

```python
# [1 ,0,0]: banane 
# [0 ,0 ,1]: pomme 
# [0,1,0]: fraise
```

# Data Preprocessing ( Scaling the images )

```python
dataset = dataset.map(lambda x,y:(x/255 , y))
```

```python
dataset.as_numpy_iterator().next()[0][1].max()
```

# Spliting the Data

```python
train_size= int(len(dataset)*.7)
val_size = int(len(dataset)*.2)+1
test_size = int(len(dataset)*.1)+1
train_size+val_size+test_size
```

```python
len(dataset)
```

```python
train = dataset.take(train_size)
val = dataset.skip(train_size).take(val_size)
test = dataset.skip(train_size+val_size).take(test_size)
```

# Build Deep Learning Model CNN #

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.layers import Lambda
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile('adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
logdir='logs'
```

```python
hist = model.fit(train,epochs=20,validation_data=val)#model.summary()
```

```python
model.summary()
```

# Plot performance

```python
fig = plt.figure()
plt.plot(hist.history['loss'],color='teal',label='loss')
plt.plot(hist.history['val_loss'],color='orange',label='val_loss')
fig.suptitle('loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

# Evaluate #

```python
from tensorflow.keras.metrics import  Accuracy
```

```python
accuracy = Accuracy()
```

```python
for batch in test.as_numpy_iterator():
    X,y= batch
    loss , acc = model.evaluate(X,y)
    print(acc)
```
